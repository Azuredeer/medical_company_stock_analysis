```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
kaef = pd.read_csv("Data Historis KAEF.csv")
kaef
```

```python
# rename kolom
kaef = kaef.rename(columns={"Tanggal" : "Date", 
                            "Terakhir" : "Close", 
                            "Pembukaan" : "Open", 
                            "Tertinggi" : "High", 
                            "Terendah" : "Low", 
                            "Vol." : "Vol", 
                            "Perubahan%" : "Change(%)"})

kaef
```

```python
# mengubah tipe data
kaef['Date'] = pd.to_datetime(kaef['Date'])

kaef['Vol'] = (kaef['Vol']
               .str.replace(',', '.', regex=True)
               .str.replace('M', 'e6', regex=True)
               .str.replace('K', 'e3', regex=True)
               .astype(float))

kaef['Change(%)'] = (kaef['Change(%)']
                     .str.replace(',', '.', regex=True)
                     .str.replace('%', '', regex=True)
                     .astype(float))
```
